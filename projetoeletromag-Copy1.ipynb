```python
from cmath import *
import numpy as np
import matplotlib.pyplot as plt
from cmath import *
from numpy import linalg
```

```python
wc=6.52e6
def impedancia(w,sigma):
    L=sigma/wc
    C=1/(sigma*wc)
    Z=(1j*w*L*(1/(1j*w*C)))/(1j*w*L+(1/(1j*w*C)))+(4*1j*w*L)
    return Z
```

```python
w=np.linspace(1e6,10e6,200)
z=[]
for i in w:
    z.append(np.imag(impedancia(i,10)))

plt.plot(w,z)
plt.ylim((-100,100))
plt.xlim((1e6,10e6))
plt.plot(wc,0,"or",color="red")
plt.plot(8.1e6,0,"or",color="blue")
plt.plot(7.29e6,0,"or",color="blue")
plt.show()    
```

```python
wr=7.9e6
M=1/wr

listaM=[]
k=0
n=1
while k <= 25:
    listaM.append(n*M)
    n-=0.02
    k+=1
```

```python
R1=1
Rc=1
G=[]
p=0
while p < len(listaM):
    Z=np.array([[R1, -1j*wr*listaM[p]],[-1j*wr*listaM[p], Rc]])
    V=np.array([1,0])
    i=np.dot(linalg.inv(Z),V)
    G.append(abs(i[1]**2/i[0]**2))
    p+=1
#G
```

```python
plt.plot(listaM, G)
plt.title("Ganho x M")
plt.ylabel('Ganho')
plt.xlabel('M')
plt.show()
```
